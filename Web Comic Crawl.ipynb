```python
!pip install roman              #install roman converter from https://github.com/zopefoundation/roman
import requests                 #to get/access/download Webpage
from bs4 import BeautifulSoup   #for crawling
import roman                    #convert roman numerals to arabic and vice versa
import shutil                   #for saving images
from PIL import Image           #for compiling images to pdf
```

```python
def getAndParseURL(url):
  getRequest = requests.get(url)
  soup = BeautifulSoup(getRequest.text, 'html.parser')
  return soup
```

```python
pics = []
url = 'https://swordscomic.com/comic/cover/'
soup = getAndParseURL(url)
image_url = soup.find("img", class_ = "comic-image").get('src')
pics.append(f"https://swordscomic.com{image_url}")
pics
```

```python
page = 1
init_page = f"https://swordscomic.com/comic/{roman.toRoman(page)}/"

while requests.get(init_page).status_code == 200:
  soup = getAndParseURL(init_page)
  image_url = soup.find("img", class_ = "comic-image").get('src')
  if image_url == '/media/404Tv2.png':
    break
  pics.append(f"https://swordscomic.com{image_url}")
  page += 1
  init_page = f"https://swordscomic.com/comic/{roman.toRoman(page)}/"
```

```python
for x in pics:
  r = requests.get(x, stream=True)
  if r.status_code == 200:
    path = f'/content/gdrive/MyDrive/Project/SwordComicsCrawl/SwordComics{pics.index(x)}.png'
    with open(path, 'wb') as f:
        r.raw.decode_content = True
        shutil.copyfileobj(r.raw, f)  
```

```python
il =[]

for x in range(len(pics)):
  i = Image.open(r'/content/gdrive/MyDrive/Project/SwordComicsCrawl/SwordComics{}.png'.format(x))
  ic = i.convert('RGB')
  il.append(ic)

img1 = Image.open('/content/gdrive/MyDrive/Project/SwordComicsCrawl/SwordComics0.png').convert('RGB')
img1.save(r'/content/gdrive/MyDrive/Project/SwordComicsCrawl/SwordComics.pdf',save_all=True, append_images=il[1:])
```
